```python
# 음료와 디저트 주문 개수 선택

# 랜덤 주문 문장 선택
order_quantity = ["한", "두", "세", "네", "다섯", "여섯", "1", "2", "3", "4", "5", "6"]
order_phrases = ["주세요.", "해줘.", "줘.", "줄래?", "부탁할게"]
```

### 단일 음료 주문

```python
import random

# 주문 문장 생성 함수
def generate_drinks():


    # 주문 생성
    order = []

    drink = random.choice(drinks)
    drink_quantity = random.choice(order_quantity)
    
    if drink_quantity in [1, 2, 3, 4, 5, 6]:
        order.append(f"{drink} {drink_quantity}잔 {random.choice(order_phrases)}")
    else:
        order.append(f"{drink} {drink_quantity} 잔 {random.choice(order_phrases)}")
    
    
    # 주문에 따른 함수

    functions = menu_dict[drink].replace("num",str(num_dict[drink_quantity]))
    
    
    return (' '.join(order), functions)
```

## 단일 메뉴 주문

```python
# 주문 문장 생성 함수
def generate_desserts():

    order = []

    # 랜덤 주문 문장 선택
    dessert = random.choice(desserts)
    dessert_quantity = random.choice(order_quantity)
    if dessert_quantity in [1, 2, 3, 4, 5, 6]:
        order.append(f"{dessert} {dessert_quantity}개 {random.choice(order_phrases)}")
    else:
        order.append(f"{dessert} {dessert_quantity} 개 {random.choice(order_phrases)}")

    # 주문에 따른 함수
    functions = menu_dict[dessert].replace("num",str(num_dict[dessert_quantity]))

    return (' '.join(order), functions)
```

## 디저트 + 디저트 주문

```python
# 주문 문장 생성 함수 수정
def generate_dessert_order_over2():
    # 함수
    functions = ""

    # 디저트 메뉴 선택
    selected_menus = random.sample(desserts, 2)

    # 주문 생성
    order = []
    for i, dessert in enumerate(selected_menus):
        quantity = random.choice(order_quantity)
        connector = random.choice(connectors) if i > 0 else ''

        if quantity in [1, 2, 3, 4, 5, 6]:
            order.append(f"{connector} {dessert} {quantity}개")
        else:
            order.append(f"{connector} {dessert} {quantity} 개")
        
        # 함수 생성
        functions = functions + menu_dict[dessert].replace("num",str(num_dict[quantity]))

    # 주문 문장 결합
    order_text = f"{''.join(order).strip()} {random.choice(tones)}"

    return (order_text, functions)
```

## 음료 + 음료 주문

```python
# 주문 문장 생성 함수 수정
def generate_drink_order_over2():
    # 함수
    functions = ""
    
    # 디저트 메뉴 선택
    selected_drinks = random.sample(drinks, 2)

    # 주문 생성
    order = []
    for i, drink in enumerate(selected_drinks):
        quantity = random.choice(order_quantity)
        connector = random.choice(connectors) if i > 0 else ''
        if quantity in [1, 2, 3, 4, 5, 6]:
            order.append(f"{connector} {drink} {quantity}개")
        else:
            order.append(f"{connector} {drink} {quantity} 개")
        
        # 함수 생성
        functions = functions + menu_dict[drink].replace("num",str(num_dict[quantity]))

    # 주문 문장 결합
    order_text = f"{''.join(order).strip()} {random.choice(tones)}"

    return (order_text, functions)
```

## 음료 + 디저트

```python
# 음료와 디저트 주문을 결합하는 함수
def generate_combined_order1():

    # 음료 주문 생성
    drink = random.choice(drinks)
    drink_quantity = random.choice(order_quantity)
    drink_order = f"{drink} {drink_quantity}잔" if drink_quantity.isdigit() else f"{drink} {drink_quantity} 잔"

    # 디저트 주문 생성
    dessert = random.choice(desserts)
    dessert_quantity = random.choice(order_quantity)
    dessert_order = f"{dessert} {dessert_quantity}개" if dessert_quantity.isdigit() else f"{dessert} {dessert_quantity} 개"

    # 전체 주문 문장 생성
    combined_order = f"{drink_order}{random.choice(connectors)} {dessert_order} {random.choice(order_phrases)}"

    # 함수
    functions = ""
    
    functions = functions + menu_dict[drink].replace("num",str(num_dict[drink_quantity]))
    functions = functions + menu_dict[dessert].replace("num",str(num_dict[dessert_quantity]))
        
    return combined_order, functions
```

##  디저트 + 음료 주문

```python
# 음료와 디저트 주문을 결합하는 함수
def generate_combined_order2():

    # 음료 주문 생성
    drink = random.choice(drinks)
    drink_quantity= random.choice(order_quantity)
    drink_order = f"{drink} {drink_quantity}잔" if drink_quantity.isdigit() else f"{drink} {drink_quantity} 잔"

    # 디저트 주문 생성
    dessert = random.choice(desserts)
    dessert_quantity = random.choice(order_quantity)
    dessert_order = f"{dessert} {dessert_quantity}개" if dessert_quantity.isdigit() else f"{dessert} {dessert_quantity} 개"

    # 전체 주문 문장 생성
    combined_order = f"{dessert_order}{random.choice(connectors)} {drink_order} {random.choice(order_phrases)}"

    # 함수
    functions = ""
    
    functions = functions + menu_dict[dessert].replace("num",str(num_dict[dessert_quantity]))
    functions = functions + menu_dict[drink].replace("num",str(num_dict[drink_quantity]))

    return combined_order, functions
```
